Initialisation du premier url du site

```python
from bs4 import BeautifulSoup
import requests

#Premier URL
url = "https://www.property24.com/for-sale/cape-town/western-cape/432?PropertyCategory=House%2cApartmentOrFlat%2cTownhouse#113562419"

#Liste qui contient tout les données finals
data = []
```

```python
def scrape(conteneur, data):
    if page.status_code == 200 :
        #conteneur des articles
        conteneur = conteneur[-1]

        #liste de tout les articles de la page
        contents = conteneur.find_all(class_="p24_content")

        for content in contents:
            dict = {'Prix': '', 'Déscription': '', 'Adresse': '', 'Chambre': '', 'Douche': '', 'Parking': '', 'Appartement(m²)': '', 'Jardin(m²)': '', 'Commentaire': ''}

            #prix
            prix = content.find(class_="p24_price").text.strip()
            print(prix)

            #déscription
            description = content.find(class_="p24_description")

            try :
                description = description.find("span").text
            except :
                description = content.find(class_="p24_title").text.strip()
                location = content.find("span", class_="p24_location").text.strip()
                description = description + " in " + location
            print(description)

            #Adresse de l'appartement
            adresse = content.find("span", class_="p24_address")

            try :
              adresse = adresse.text.strip()
            except :
              adresse = None

            print(f"L'adresse de l'appartement : {str(adresse)}")

            #nombre de chambre
            chambre = content.find("span", class_="p24_featureDetails", title="Bedrooms")

            try :
                chambre = chambre.find("span").text.strip()
            except :
                chambre = None

            print(f"Chambre : {str(chambre)}")

            #nombre de douche
            douche = content.find("span", class_="p24_featureDetails", title="Bathrooms")

            try :
                douche = douche.find("span").text.strip()
            except :
                douche = None

            print(f"Douche : {str(douche)}")

            #nombre de parking
            parking = content.find("span", class_="p24_featureDetails", title="Parking Spaces")

            try :
                parking = parking.find("span").text.strip()
            except :
                parking = None
            print(f"Parking : {str(parking)}")

            #périmètre de l'appartement
            perim_appart = content.find("span", class_="p24_size", title="Erf Size")

            try :
                perim_appart = perim_appart.find("span").text.strip()
            except :
                perim_appart = None
            print(f"Appart(m²) : {str(perim_appart)}")

            #périmètre du jardin
            perim_jardin = content.find("span", class_="p24_size", title="Floor Size")

            try :
                perim_jardin = perim_jardin.find("span").text.strip()
            except :
                perim_jardin = None

            print(f"jardin(m²) : {str(perim_jardin)}")

            #commentaire pour l'appartement
            comment = content.find("span", class_="p24_excerpt")

            try :
                comment = comment.text.strip()
            except :
                comment = None
            print(f"Commentaire : {str(comment)}\n")

            #On récolte tout les donnée dans une dictionnaire
            dict = {
                'Prix': prix,
                'Déscription': description,
                'Adresse': adresse,
                'Chambre': chambre,
                'Douche': douche,
                'Parking': parking,
                'Appartement(m²)': perim_appart,
                'Jardin(m²)': perim_jardin,
                'Commentaire': comment
            }
            #On l'ajoute dans un liste finale
            data.append(dict)

    else:
        print(f"Une erreur est survenue(code d'erreur : {page.status_code})")

```

Parcours pour tout les pages du site

```python
i = 1
while True:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    conteneur = soup.find_all(class_="js_listingResultsContainer")

    if conteneur == []:
      break
    else :
      scrape(conteneur, data)
      print(f"Scrape de la page {str(i)} términer")
      i += 1
      page_suivant = f"https://www.property24.com/for-sale/cape-town/western-cape/432/p{str(i)}?PropertyCategory=House%2cApartmentOrFlat%2cTownhouse#113562419"
      url = page_suivant
```
