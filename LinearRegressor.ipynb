```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
class LinearRegression:
    def __init__(self, X, y, b=True):
        self.y = y
        self.b = b
        if b == True:
            self.X = np.hstack((X, np.ones((X.shape[0], 1))))
        else:
            self.X = X

    def train(self):
        self.y = self.y.reshape((self.y.shape[0], 1))
        return np.linalg.inv(self.X.T.dot(self.X)).dot(self.X.T.dot(self.y))

    def predict(self):
        prediction = self.X.dot(self.train())

        return prediction
```

```python
X, y = make_regression(n_samples=100, n_features=5, n_targets=1)
```

```python
linear = LinearRegression(X, y, b=True)
```

```python
linear.predict()[:10]
```

```python
class RidgeRegression:
    def __init__(self, X, y, lbda,b=True):
        self.y = y
        self.b = b
        self.lbda = lbda
        if b == True:
            self.X = np.hstack((X, np.ones((X.shape[0], 1))))
        else:
            self.X = X

    def train(self):
        self.y = self.y.reshape((self.y.shape[0], 1))
        return np.linalg.inv(self.X.T.dot(self.X) + (self.lbda * np.eye(self.X.shape[1]))).dot(self.X.T.dot(self.y))

    def predict(self):
        prediction = self.X.dot(self.train())

        return prediction
```

```python
ridge = RidgeRegression(X, y, 0.25, b=True)
```

```python
ridge.train()
```

```python
ridge.predict()[:10]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
trovit = pd.read_csv('trovit.csv')
rawson = pd.read_csv('rawson.csv')
pamgold = pd.read_csv('pamgolding.csv')
sahome = pd.read_csv('sahometraders.csv')
area = pd.read_csv('area.csv')
covid = pd.read_csv('covid_19_clean_complete.csv')
```

```python
trovit = trovit['Price']
rawson = rawson['Price']
pamgold = pamgold['Price']
sahome = sahome['Price']
area = area['Price']
covid = covid['Deaths']
```

```python
def countInt(X):
    X = X.dropna(axis=0)
    nombre = ['1', '2', '3', '4', '5', '6', '7', '8', '9']
    liste = []
    for value in X:
        str_value = str(value)
        for i in range(len(str_value)):
            if str_value[i] not in nombre:
                continue
            else:
                liste.append(int(str_value[i]))
                break

    plt.figure()
    sns.histplot(liste)
```

```python
countInt(trovit)
```

```python
countInt(sahome)
```

```python
countInt(area)
```
